#####  Identifying and Eliminating outliers

```python
import matplotlib.pyplot as plt
```

```python
adv=pd.read_csv("Advertising.csv")
adv.head(11)
```

```python
adv.plot(kind="box")
plt.show()
```

```python
Q3=adv.quantile(0.75)
Q1=adv.quantile(0.25)
IQR=Q3-Q1
IQR
```

```python
fliter_data=adv[((adv<(Q1-1.5*IQR))|(adv>(Q3+1.5*IQR))).any(axis=1)]
fliter_data
```

```python
fliter_data=adv [~((adv<(Q1 - 1.5 * IQR))|(adv>(Q3 + 1.5 * IQR))).any(axis=1)]
fliter_data
```

```python
fliter_data.plot(kind="box")
plt.show()
```
