### matplotlib
- matplotlib is one the data science package
-it is a cross platform data visulization library for making 2d plots from data

#### types of plotting
- .lines plot
- .bar graph
- .scatter plot
- .area plot
- .pie charts

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[4,2,8,10],"r")
plt.title("xy plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
marks=[95,20,85,65,80]
subject=["python","django","html","css","data analysis"]
plt.bar(subject,marks,color=["r","b","y","green","black"],width=0.9)
plt.show()
```

```python
#scatter
plt.scatter([2,4,6,8],[4,2,6,3])
plt.title("xy scatter plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# area plot
days=[1,2,3,4]
work=[10,12,8,6]
sleeping=[8,8,12,11]
plt.stackplot(days,work,sleeping,colors=["black","blue"],labels=["work","rest"])
plt.legend()
plt.show()
```

```python
# pie chart
marks=[25,15,30,24,6]
subjects=["python","data","html","css","django"]
plt.pie(marks,colors=["b","g","black","y","r"],labels=subjects)
plt.legend()
plt.show()
```

```python
import numpy as np
x=np.arange(3,20)
y=2*x+8
plt.plot(x,y,"r")
plt.title("xy grapy using numpy")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
